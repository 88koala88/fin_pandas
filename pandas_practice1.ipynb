### iris 데이터 로드

```python
# pd.read_csv
df = pd.read_csv("iris.csv")
```

```python
df
```

컬럼 출력하기

```python
df['Sepal.Length']
```

컬럼 데이터프레임 형식으로 출력하기

```python
df[['Sepal.Length']]
```

```python
df[['Sepal.Length', 'Species']]
```

컬럼 조건문으로 필터링하기
- 컬럼을 조건을 달아서 필터링하기 위해선 단순히 []안에 컬럼명을 집어넣는것이 아니라 (boolean 값으로 출력이됨) 컬럼안에 다시 한번 []를 집어넣어 데이터 프레임 형식으로 출력을 해줘야 우리가 원하는 형태로 출력이 된다

```python
df['Sepal.Length'] > 5
```

```python
df[df['Sepal.Length'] > 5]
```

컬럼 간의 관계 조건식으로 출력을 하고 싶을 경우에는 반드시 () 괄호 처리를 해줘야 한다.

```python
# 조건식
# & | 
# AND, OR
# == 
# != 
```

```python
df[
    (df['Sepal.Length'] > 5) & 
    (df['Sepal.Width'] > 3)
]
```

```python
df[(df['Sepal.Length'] > 5) & (df['Sepal.Width'] > 3) & (df['Species'] == 'virginica')]
```

```python
df1 = df[(df['Sepal.Length'] > 5) & (df['Sepal.Width'] > 3) & (df['Species'] == 'virginica')]
```

데이터 프레임을 전처리 후 출력할 때 항상 인덱스 정리에 신경을 써주자

```python
df1.reset_index()
```

인덱스를 정리해줄 때 drop을 사용하여 인덱스항을 정리해주자

```python
df1.reset_index(drop = True)
```

```python
df2 = df1.reset_index(drop = True)
```

데이터 프레임을 파일로 저장할 때 항상 index = false 를 잊지말자!

```python
df2.to_csv('virginica.csv', index = False)
```

```python
pd.read_csv('virginica.csv')
```

drop_duplicates() 메서드를 사용하면 중복값을 제거하고 컬럼의 대표항목만 추려낸다.

```python
df[['Species']].drop_duplicates()
```

groupby 메서드로 원하는 항목을 기준으로 그룹화 하여 각종 항목값을 추려낼 수 있다.

```python
df.groupby('Species').sum()
```

```python
df2 = df.groupby('Species').sum()
```

```python
df3 = df.groupby('Species').mean()
```
