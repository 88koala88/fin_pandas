데이터 로드

```python
AptRent_Total = pd.read_csv("data/AptRent_Total.csv")
```

필요없는 컬럼 제거하기

```python
AptRent_Total = AptRent_Total.drop(['갱신요구권사용','계약구분','종전계약보증금','종전계약월세'], axis=1)
```

계산식 활용하여 새로운 컬럼 생성해주기

```python
AptRent_Total['연식'] = 2023 - AptRent_Total['건축년도']
```

```python
AptRent_Total.head()
```

조건식 활용하여 컬럼 필터링

```python
# 연식이 10년 미만인
AptRent_Total[AptRent_Total['연식'] < 10]
```

```python
# 전용 면적이 84 이상인
AptRent_Total[AptRent_Total['전용면적'] >= 84]
```

```python
# 보증금액이 50000 이상인
AptRent_Total[AptRent_Total['보증금액'] >= 50000]
```

```python
# 보증금액 50000 이하 동시에 월세금액 0이 아닌 동시에 월세금액 100
AptRent_Total[(AptRent_Total['보증금액'] <= 50000) & (AptRent_Total['월세금액'] != 0) & (AptRent_Total['월세금액'] == 100)]
```

```python
AptRent_Total[AptRent_Total['법정동'] == '반포동']
```

groupby 메서드 활용하여 데이터 필터링

```python
AptRent1 = AptRent_Total[['보증금액','월세금액','법정동']].groupby(['법정동']).mean()
```

```python
AptRent2 = AptRent2.reset_index()
```

sort_values 활용하여 데이터 정렬하기 (ascending = False로 지정할 경우 내림차순 정렬)

groupby와 각종 조건식 활용하여 데이터 전처리 실습

```python
# 연식별 보증금액
AptRent_Total[['보증금액','연식']].groupby(['연식']).mean()
```

```python
# 법정동별 보증금액
apt1 = AptRent_Total[['보증금액','법정동']].groupby(['법정동']).mean()
apt2 = apt1.reset_index()
apt2.sort_values(['보증금액'], ascending = False)
```

```python
# 월별, 법정동별 보증금액
# 보증금액이 높은것기준 정렬
Apt1 = AptRent_Total[['보증금액','법정동','월']].groupby(['법정동','월']).mean()
Apt2 = Apt1.reset_index()
Apt2.sort_values(['보증금액'], ascending = False)
```

```python
# 월세금액이 0인것 (전세)
# 법정동, 연식별 보증금액
apt1 = AptRent_Total[AptRent_Total['월세금액'] == 0]
apt1 = apt1[['보증금액','법정동','연식']].groupby(['법정동','연식']).mean()
apt2 = apt1.reset_index()
apt2.sort_values(['보증금액'], ascending = False)
```

```python
# 월세금액이 0이 아닌것 (월세)
# 법정동, 연식별 보증금액, 월세금액
apt1 = AptRent_Total[AptRent_Total['월세금액'] != 0]
apt1 = apt1[['보증금액','법정동','연식','월세금액']].groupby(['법정동','연식']).mean()
apt2 = apt1.reset_index()
apt2.sort_values(['월세금액'], ascending = False)
```
